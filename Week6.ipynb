# Exploratory Data Analysis (EDA) on Data Science Jobs Dataset
## TDI Python for Data Analytics - Week 6
Dataset: `dsjobs_data.csv`

This notebook performs data cleaning, exploratory data analysis, and visualization on the dataset.



```python
# Load required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set style for seaborn
sns.set(style="whitegrid")

# Load dataset
file_path = "dsjobs_data.csv"
df = pd.read_csv(file_path)

# Display the first 5 rows
df.head()
```

# 1. Inspect the Data Structure
# Examine the dataset for an overview of the columns, data types, and missing values.

```python
df.info()
df.describe(include='all')
```

# 2. Handle Missing Values
# Identify columns with missing values and decide on appropriate handling methods.

```python
df.isnull().sum()
```

# 3. Extract Salary Information
# Extract min and max salary from 'Salary Estimate' column and calculate the average salary.

```python
df['Salary Estimate'] = df['Salary Estimate'].astype(str)
df['Salary Estimate'] = df['Salary Estimate'].str.replace(r'\D', '', regex=True)
df['Min Salary'] = df['Salary Estimate'].str.extract('(\d{2,5})').astype(float)
df['Max Salary'] = df['Salary Estimate'].str.extract('(\d{2,5})$', expand=True).astype(float)
df['Average Salary'] = (df['Min Salary'] + df['Max Salary']) / 2
df[['Min Salary', 'Max Salary', 'Average Salary']].head()
```

# 4. Standardize Categorical Data
# Clean and standardize 'Company Name' and 'Location'.

```python
df['Company Name'] = df['Company Name'].str.replace('\n', '', regex=True).str.strip()
df['Location'] = df['Location'].str.strip()
df[['Company Name', 'Location']].head()
```

# 5. Correct Data Types and Drop Duplicates
# Ensure proper data types and remove duplicate rows.

```python
df = df.drop_duplicates()
df.info()
```

# 6. Summary Statistics
# Calculate basic descriptive statistics for 'Rating', 'Average Salary', and 'Revenue'.

```python
df[['Rating', 'Average Salary', 'Revenue']].describe()
```

# 7. Categorical Overview
# Determine the top 5 most common job titles and their average ratings.

```python
df.groupby('Job Title')['Rating'].mean().nlargest(5)
```

# 8. Top 3 industries with the highest average company rating

```python
df.groupby('Industry')['Rating'].mean().nlargest(3)
```

# 9. Company Size Analysis
# Analyze the distribution of company sizes.

```python
df['Size'].value_counts()
```

# 10. Determine the average salary range for each company size category

```python
df.groupby('Size')['Average Salary'].mean()
```

# 11. Salary Distribution
# Create a histogram for 'Average Salary'.

```python
plt.figure(figsize=(8, 5))
plt.hist(df['Average Salary'].dropna(), bins=20, color='blue', edgecolor='black', alpha=0.7)
plt.xlabel('Average Salary')
plt.ylabel('Frequency')
plt.title('Salary Distribution')
plt.show()
```

# 12. Company Ratings by Location
# Bar chart comparing average company ratings across top 10 locations.

```python
top_locations = df['Location'].value_counts().nlargest(10).index
avg_ratings = df[df['Location'].isin(top_locations)].groupby('Location')['Rating'].mean()

plt.figure(figsize=(10, 5))
avg_ratings.sort_values().plot(kind='bar', color='green')
plt.xlabel('Location')
plt.ylabel('Average Rating')
plt.title('Company Ratings by Location')
plt.show()
```

# 13. Company Size vs. Salary
# Line plot showing the average salary for each company size category.

```python
avg_salary_by_size = df.groupby('Size')['Average Salary'].mean()
plt.figure(figsize=(8, 5))
avg_salary_by_size.plot(marker='o', linestyle='-')
plt.xlabel('Size')
plt.ylabel('Average Salary')
plt.title('Size vs. Salary')
plt.xticks(rotation=45)
plt.grid()
plt.show()
```

# 14. Correlation Heatmap
# Create a heatmap of correlations between numerical features.

```python
plt.figure(figsize=(8, 6))
sns.heatmap(df[['Rating', 'Average Salary', 'Founded']].corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```

# 15. Job Title and Rating Analysis
# Box plot for ratings of the top 5 job titles.

```python
top_jobs = df['Job Title'].value_counts().nlargest(5).index
plt.figure(figsize=(10, 5))
sns.boxplot(x='Job Title', y='Rating', data=df[df['Job Title'].isin(top_jobs)])
plt.xticks(rotation=45)
plt.title('Job Title and Rating Analysis')
plt.show()
```

# 16. Salary Comparison by Industry
# Bar plot showing average salary for the top 10 industries.

```python
top_industries = df['Industry'].value_counts().nlargest(10).index
avg_salary_by_industry = df[df['Industry'].isin(top_industries)].groupby('Industry')['Average Salary'].mean()

plt.figure(figsize=(10, 5))
avg_salary_by_industry.sort_values().plot(kind='barh', color='purple')
plt.xlabel('Average Salary')
plt.ylabel('Industry')
plt.title('Salary Comparison by Industry')
plt.show()
```

# 17. Pair Plot of Key Metrics
# Create a pair plot to visualize relationships between 'Rating', 'Founded' year, and 'Average Salary'.

```python
sns.pairplot(df[['Rating', 'Founded', 'Average Salary']].dropna());
```
